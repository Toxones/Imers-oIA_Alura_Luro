definindo uma função para que o gemini retorne os dados embbedados dos documentos - utilizando o task type Retrieval_Document que realiza a busca de dados em documentos transformando tudo que está documentado em vetores.

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             title=title,
                             task_type='RETRIEVAL_DOCUMENT')['embedding']
```

Vinculando os vetores emmbedados de cada documento com a sua coluna respectiva no dataframe.

```python
df['Embeddings'] = df.apply(lambda row: embed_fn(row['TITULO'], row['CONTEUDO']), axis=1)
df
```

Linha 1~4 - definindo a consulta que será utilizada para fazer a busca de informação na base de dados do modelo.

Linha 5 - definindo a função que será utilizada para realizar a busca da informação mais próxima do que for colocado no prompt de input - calculando a diferença vetorial entre os dois pontos palavra por palavra.

Linha 7~8 - definindo o indice como o maior argumento encontrado na consulta de produtos escalares, fazendo com que a consulta retorne o titulo baseado nesse resultado do indice.

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                             content=consulta,
                            task_type='RETRIEVAL_QUERY')
  produtos_escalares = np.dot(np.stack(df['Embeddings']), embedding_da_consulta['embedding'])

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]['TITULO']
```

Configurações de acuracia do modelo generativo do Gemini e definindo segundo modelo para trabalharmos com uma resposta melhor ao usuário

Startando um modelo generativo de chat e solicitando que ele grave o histórico da conversa.

Criando a consulta como um input do chat, definindo ot recho como a resposta a função a partir da consulta, definindo o prompt para que o gemini aja como consultor da Alura
Definindo um laço de while para que seja uma conversa continua entre o usuário e o sistema.

```python
consulta = input('O que você gostaria de estudar na alura?' '\n')
trecho = gerar_e_buscar_consulta(consulta, df, model)
prompt = f'aja como um consultor da Alura e além de dizer o nome da escola também de mais informações que fazem parte do conteudo da escola escolhida, sem adicionar informações que não estão no conteudo, ofereça um desconto de 10% para fechar um plano nessa escola: {trecho}'
while consulta !='fim':
  response = model_2.generate_content(prompt)
  print('\n\n', response.text, '\n\n')
  consulta = input('o que você gostaria de estudar na alura?' '\n')
```
